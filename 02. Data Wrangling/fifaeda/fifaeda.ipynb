# Зависит ли Стоимость игрока стоимость (Value) игрока от его "финтов" (Skill Moves)? 

```python
def Converter(x):
  x = str(x)[1:]
  if len(x) != 1:
    if x[-1] == 'M':
      return float(x[:-1]) * 1000000
    return float(x[:-1]) * 1000
  return 0
```

```python
skill_value = df.groupby('Skill Moves')['Value'].value_counts().reset_index(name="Count")
skill_value['Double_Value'] = pd.Series(list(map(Converter, skill_value['Value'])))
skill_value = skill_value.sort_values(by='Double_Value')
fig = go.Figure()
fig = px.histogram(skill_value.sort_values(by='Double_Value'), x = 'Value', y ='Count', color="Skill Moves")
fig.update_layout(title='Number of players with each skill by value',
                  xaxis_title="Value",
                  yaxis_title="Number of players",
                  barmode='stack')
fig.update_layout(legend_traceorder="normal")
fig.show(renderer='colab')
```

Явная зависимость: чем больше Skill, тем больше стоимость игрока.

# Найдите суммарную стоимость игроков в каждом клубе. Визуализируйте соотношение сумарной стоимости топ 10 клубов. 

```python
def Converter2(x):
  x = str(x)[1:]
  if x[-1] == "M":
    return float(x[:-1])* 1000000
  if x[-1] == "K":
    return float(x[:-1])* 1000
  return float(x)
```

```python
df_total_value = df.groupby('Club').apply(lambda x: round(sum(x["Value"].apply(Converter2)), 1)).reset_index(name='Total Value')
fig = go.Figure()
fig.add_trace(go.Bar(x = df_total_value['Club'].head(10), y = df_total_value['Total Value'].head(10)))
fig.update_layout(title='The total cost of the club\'s players',
                  xaxis_title="Club",
                  yaxis_title="Cost €", 
                  bargap=0.2)
fig.show(renderer='colab')
```

# Есть ли зависимость между продолжительностью контракта игрока и его стоимость? 

```python
def convert_wage(s):
    if s[-1] == 'K':
        return float(s[1:len(s) - 1]) * 1000
    else:
        return float(s[1:])
def convert_year(s):
    if type(s) == float:
        return s
    else:
        return int(s[-4:]) 
```

```python
df['Time in Club'] = df['Contract Valid Until'].apply(convert_year) - df['Joined'].apply(convert_year)
df['Wage Float'] = df['Wage'].apply(convert_wage)
df_wage_year = df.groupby('Time in Club').mean()['Wage Float']
fig = px.line(df_wage_year,
              x = df_wage_year.index,
              y = 'Wage Float',
              title = 'Dependence mean wage by contract duration')
fig.show(renderer='colab')
```

Чем дольше длится контракт, тем больше зп до 16. Но, начиная с 16 зп становится меньше.

# Насколько средняя зарплата (Wage) игроков с рейтингом (Overall) 90+ больше средней зарплаты с рейтингом [80 - 90]? Визуализируйте возможную зависимость

```python
df_wage_by_overall = df.groupby('Overall').mean()['Wage Float'].reset_index(name='Wage Float')
mean_wage_big_overall = df_wage_by_overall[df_wage_by_overall['Overall']>90].mean()[1]
mean_wage_small_overall = df_wage_by_overall[(df_wage_by_overall['Overall']<=90) & (df_wage_by_overall['Overall']>=80)].mean()[1]
print(f"Разница в средней зарплате равна {mean_wage_big_overall - mean_wage_small_overall}€")
fig = go.Figure()
fig.add_trace(go.Bar(x = ["80-90"], y = [mean_wage_small_overall]))
fig.add_trace(go.Bar(x = ["90+"], y = [mean_wage_big_overall]))
fig.update_layout(title='Mean wage by overral',
                  xaxis_title="Overall",
                  yaxis_title="Mean wage €", 
                  bargap=0.2, showlegend=False)
fig.show(renderer='colab')
```
